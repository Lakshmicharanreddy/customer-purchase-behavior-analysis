```python
#TOTAL CUSTOMERS

query = """

SELECT COUNT(DISTINCT CUSTOMER_ID) AS TOTAL_CUSTOMERS

FROM CUSTOMERS;

"""

total_customers = pd.read_sql(query,engine)


total_customers
```

```python

#TOTAL ORDERS
query = """

SELECT COUNT(DISTINCT ORDER_ID) AS TOTAL_ORDERS

FROM ORDERS;

"""

total_orders = pd.read_sql(query,engine)


total_orders
```

```python
#TOTAL REVENUE

query = """
SELECT SUM(PAYMENT_VALUE) AS TOTAL_REVENUE

FROM ORDER_PAYMENTS;
"""

total_revenue = pd.read_sql(query,engine)


total_revenue
```

```python
#AVERAGE ORDER VALUE

query = """

SELECT SUM(OP.PAYMENT_VALUE) / COUNT(DISTINCT O.ORDER_ID) AS AVERAGE_ORDER_VALUE

FROM ORDER_PAYMENTS OP
JOIN ORDERS O
ON OP.ORDER_ID = O.ORDER_ID;


"""

average_order_value = pd.read_sql(query,engine)


average_order_value


```

```python
#ORDER STATUS DISTRIBUTION


query = """

SELECT ORDER_STATUS,
       COUNT(ORDER_ID) AS TOTAL_ORDERS


FROM ORDERS 
GROUP BY ORDER_STATUS;

"""

status_df = pd.read_sql(query,engine)


status_df
```

```python
ax = sns.barplot(data = status_df,x = 'order_status', y = 'total_orders',hue = "total_orders",legend = False)

for bar in ax.containers:
    ax.bar_label(bar)


plt.title("ORDER STATUS DISTRIBUTION")
plt.xlabel("Order Status")
plt.ylabel("Total Orders")
plt.show()

    
```

```python
# Average Revenue per Month



query = """


SELECT DATE_TRUNC('MONTH',O.ORDER_PURCHASE_TIMESTAMP::TIMESTAMP) AS ORDER_MONTH,
       AVG(OP.PAYMENT_VALUE) AS AVERAGE_REVENUE


FROM ORDERS O
JOIN ORDER_PAYMENTS OP
ON O.ORDER_ID = OP.ORDER_ID

GROUP BY DATE_TRUNC('MONTH',O.ORDER_PURCHASE_TIMESTAMP::TIMESTAMP)
ORDER BY ORDER_MONTH;



"""

monthly_average_revenue_df = pd.read_sql(query,engine)

monthly_average_revenue_df



```

```python
#DAILY ORDERS 

query = """


WITH DOW_ORDERS AS (

SELECT EXTRACT(DOW FROM ORDER_PURCHASE_TIMESTAMP::TIMESTAMP) AS DAY_OF_THE_WEEK,
	   COUNT(ORDER_ID) AS TOTAL_ORDERS


FROM ORDERS 

GROUP BY EXTRACT(DOW FROM ORDER_PURCHASE_TIMESTAMP::TIMESTAMP)
ORDER BY DAY_OF_THE_WEEK

)
SELECT WEEK,
	   TOTAL_ORDERS

FROM (

SELECT *,
CASE
	WHEN DAY_OF_THE_WEEK = 0 THEN 'SUNDAY'
	WHEN DAY_OF_THE_WEEK = 1 THEN 'MONDAY'
	WHEN DAY_OF_THE_WEEK = 2 THEN 'TUESDAY'
	WHEN DAY_OF_THE_WEEK = 3 THEN 'WEDNESDAY'
	WHEN DAY_OF_THE_WEEK = 4 THEN 'THURSDAY'
	WHEN DAY_OF_THE_WEEK = 5 THEN 'FRIDAY'
	ELSE 'SATURDAY'
END AS WEEK

FROM DOW_ORDERS

)


"""

weekly_orders_df = pd.read_sql(query,engine)

weekly_orders_df 


```

```python
ax = sns.barplot(data = weekly_orders_df,x = 'week', y = 'total_orders' , hue = 'total_orders',legend = False)

for bar in ax.containers:
    ax.bar_label(bar)


plt.title("DAY OF THE WEEK TOTAL ORDERS")
plt.xlabel("WEEK DAY")
plt.ylabel("Total Orders")
plt.xticks(rotation = 45)
plt.show()


```

```python
#CUSTOMER TYPE



query = """


SELECT CUSTOMER_TYPE,
	   COUNT(*) TOTAL_CUSTOMERS

FROM (

SELECT 
CASE
	WHEN COUNT(O.ORDER_ID) = 0 THEN 'NO ORDERS'
	WHEN COUNT(O.ORDER_ID) = 1 THEN 'ONE-TIME CUSTOMERS'
	ELSE 'REPEAT CUSTOMERS'
END AS CUSTOMER_TYPE

FROM CUSTOMERS C
LEFT JOIN ORDERS O
ON C.CUSTOMER_ID = O.CUSTOMER_ID

GROUP BY C.CUSTOMER_ID

) SUB


GROUP BY CUSTOMER_TYPE;



"""

customer_type_df = pd.read_sql(query,engine)

customer_type_df







```

```python


plt.figure(figsize=(6,6))
plt.pie(customer_type_df['total_customers'],
        labels=customer_type_df['customer_type'],
        autopct='%1.1f%%')

plt.title("Repeat vs One-Time Customers")
plt.show()
```

```python
#Average Product Price by Category




query = """


SELECT PC.PRODUCT_CATEGORY_NAME_ENGLISH,
	   AVG(OI.PRICE) AS AVG_PRODUCT_CATEGORY_PRICE



FROM PRODUCT_CATEGORY PC
JOIN PRODUCTS P
ON PC.PRODUCT_CATEGORY_NAME = P.PRODUCT_CATEGORY_NAME
JOIN ORDER_ITEMS OI
ON P.PRODUCT_ID = OI.PRODUCT_ID

GROUP BY PC.PRODUCT_CATEGORY_NAME_ENGLISH

ORDER BY AVG_PRODUCT_CATEGORY_PRICE DESC

LIMIT 10;



"""

avg_product_price_by_category_df = pd.read_sql(query,engine)

avg_product_price_by_category_df  

```

```python

plt.figure(figsize = (10,6))
sns.barplot(data = avg_product_price_by_category_df    , x= 'product_category_name_english' , y= 'avg_product_category_price', hue = 'avg_product_category_price', legend = False)


plt.title("TOP Products Categories Average Price Distribution")
plt.xlabel("Category")
plt.ylabel("Avg_Product_Price")
plt.xticks(rotation = 90)
plt.show()


```

```python
#Top Sellers by Revenue



query = """

SELECT S.SELLER_ID,
	   ROUND(SUM(OP.PAYMENT_VALUE)::NUMERIC,2) AS TOTAL_REVENUE


FROM SELLERS S
JOIN ORDER_ITEMS OI
ON S.SELLER_ID = OI.SELLER_ID
JOIN ORDER_PAYMENTS OP
ON OI.ORDER_ID = OP.ORDER_ID

GROUP BY S.SELLER_ID
ORDER BY TOTAL_REVENUE DESC
LIMIT 10;


"""

top_sellers_by_revenue_df = pd.read_sql(query,engine)

top_sellers_by_revenue_df 

```

```python

plt.figure(figsize = (10,6))
ax = sns.barplot(data = top_sellers_by_revenue_df , x= 'seller_id' , y= 'total_revenue', hue = 'total_revenue', legend = False)

for bar in ax.containers:
    ax.bar_label(bar)


plt.title("Top Sellers By Revenue")
plt.xlabel("Sellers ID")
plt.ylabel("Revenue")
plt.xticks(rotation = 90)
plt.show()


```

```python
# DELIVERY STATUS 

query = """

WITH AVERAGE_DELIVERY_DAYS AS (

SELECT OI.SELLER_ID,
	   ROUND(AVG(
	     DATE_PART('DAY',O.ORDER_DELIVERED_CUSTOMER_DATE - O.ORDER_PURCHASE_TIMESTAMP::TIMESTAMP)
	   )::NUMERIC,2) AS AVG_DELIVERY_DAYS


FROM ORDER_ITEMS OI
JOIN ORDERS O
ON OI.ORDER_ID = O.ORDER_ID

WHERE O.ORDER_DELIVERED_CUSTOMER_DATE IS NOT NULL

GROUP BY OI.SELLER_ID
ORDER BY AVG_DELIVERY_DAYS


)

,DELIVERY_STATUS_CTE AS (

SELECT *,
CASE
	WHEN AVG_DELIVERY_DAYS <= 2 THEN 'FAST DELIVERY'
	WHEN AVG_DELIVERY_DAYS BETWEEN 3 AND 4 THEN 'MEDIUM DELIVERY'
	WHEN AVG_DELIVERY_DAYS BETWEEN 4 AND 6 THEN 'SLOW DELIVERY' 
	ELSE 'VERY SLOW DELIVERY'
END AS DELIVERY_STATUS

FROM AVERAGE_DELIVERY_DAYS

)
SELECT DELIVERY_STATUS,
	   COUNT(SELLER_ID) AS TOTAL_SELLERS


FROM DELIVERY_STATUS_CTE

GROUP BY DELIVERY_STATUS

"""


delivery_status_df = pd.read_sql(query,engine)

delivery_status_df


```

```python


plt.figure(figsize = (10,6))
ax = sns.barplot(data =delivery_status_df , x= 'delivery_status' , y= 'total_sellers', hue = 'total_sellers', legend = False)

for bar in ax.containers:
    ax.bar_label(bar)


plt.title("Sellers by Delivery Status")
plt.xlabel("Delivery Status")
plt.ylabel("Total sellers")

plt.show()

```

```python
#DELIVERY DELAY IMPACT ON REVIEWS


query = """


SELECT O.ORDER_ID,
	   DATE_PART('DAY',O.ORDER_DELIVERED_CUSTOMER_DATE - ORDER_ESTIMATED_DELIVERY_DATE) AS DELIVERY_DELAY,
	   ORE.REVIEW_SCORE


FROM ORDERS O
JOIN ORDER_REVIEWS ORE
ON O.ORDER_ID = ORE.ORDER_ID

WHERE O.ORDER_DELIVERED_CUSTOMER_DATE IS NOT NULL



"""


delivery_delay_df = pd.read_sql(query,engine)

delivery_delay_df

```

```python

sns.scatterplot(data = delivery_delay_df, x =  'delivery_delay', y = 'review_score', alpha = 0.6,color="steelblue", edgecolor="black")

plt.title("Delivery Delay vs Review Score")
plt.xlabel("Delivery Delay (days)")
plt.ylabel("Review Score (1–5)")

sns.lmplot(x="delivery_delay", y="review_score", data= delivery_delay_df,
           scatter_kws={"alpha":0.5, "color":"steelblue"},
           line_kws={"color":"red"})

plt.show()
            
```

```python
#Average Review Score by Product Category


query = """


SELECT PC.PRODUCT_CATEGORY_NAME_ENGLISH,
	   AVG(REVIEW_SCORE) AS AVG_REVIEW_SCORE


FROM PRODUCT_CATEGORY PC
JOIN PRODUCTS P
ON PC.PRODUCT_CATEGORY_NAME = P.PRODUCT_CATEGORY_NAME
JOIN ORDER_ITEMS OI
ON P.PRODUCT_ID = OI.PRODUCT_ID
JOIN ORDER_REVIEWS R
ON OI.ORDER_ID = R.ORDER_ID


GROUP BY PC.PRODUCT_CATEGORY_NAME_ENGLISH

ORDER BY AVG_REVIEW_SCORE DESC
LIMIT 10

"""


avg_category_review_df = pd.read_sql(query,engine)

avg_category_review_df


```

```python


plt.figure(figsize = (10,6))
ax = sns.barplot(data = avg_category_review_df  , x= 'product_category_name_english',y= 'avg_review_score', hue = 'avg_review_score', legend = False)

for bar in ax.containers:
    ax.bar_label(bar)


plt.title("TOP Average Review Score by Product Category")
plt.xlabel("Product category")
plt.ylabel("AVG Review Score")
plt.xticks(rotation = 90)
plt.show()

```

```python
#REVIEWS OVER TIME

query = """


SELECT DATE_TRUNC('MONTH',REVIEW_CREATION_DATE) AS REVIEW_MONTH,
	   AVG(REVIEW_SCORE) AS AVG_REVIEW_SCORE


FROM ORDER_REVIEWS 

GROUP BY DATE_TRUNC('MONTH',REVIEW_CREATION_DATE)

ORDER BY REVIEW_MONTH


"""

reviews_over_time_df = pd.read_sql(query,engine)
reviews_over_time_df 




```

```python
sns.lineplot(data = reviews_over_time_df , x = 'review_month' , y = 'avg_review_score')

plt.title("REVIEWS OVER TIME")
plt.xlabel("review month")
plt.ylabel("avg review score") 
plt.xticks(rotation = 45)
plt.show()

             
```
